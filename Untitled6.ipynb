# 2.Does cubicinches of a vehicle affects the vehicles time to reach 60 Miles per hour?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))
sns.regplot(x='cubicinches', y='time-to-60', data=data, scatter_kws={'s': 100}, line_kws={'color': 'red', 'linewidth': 2})

plt.title('Cubic Inches vs Time to 60 MPH')
plt.xlabel('Cubic Inches')
plt.ylabel('Time to 60 MPH (seconds)')

plt.show()
```

```python
data['cubicinches'] = pd.to_numeric(data['cubicinches'], errors='coerce')
data['time-to-60'] = pd.to_numeric(data['time-to-60'], errors='coerce')

data.dropna(subset=['cubicinches', 'time-to-60'], inplace=True)

correlation = data['cubicinches'].corr(data['time-to-60'])
print(f'Correlation coefficient between cubicinches and time-to-60: {correlation}')


```

# 4.How does the horse power affects the average of the vehicle ?

```python
plt.figure(figsize=(15, 10))

# Create a scatter plot and add a regression line
sns.regplot(x='hp', y='mpg', data=data, scatter_kws={'s': 100}, line_kws={'color': 'red', 'linewidth': 2})

# Add title and labels
plt.title('Horsepower vs MPG with Trend Line')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Miles Per Gallon (mpg)')

# Show the plot
plt.show()
```

# 5. How does the cubic inches affects the average of the vehicle ?


```python
plt.figure(figsize=(15, 10))

data_sorted = data.sort_values(by='cubicinches')

plt.bar(data_sorted['cubicinches'], data_sorted['mpg'], color='b')

plt.title('Cubic Inches vs MPG')
plt.xlabel('Cubic Inches')
plt.ylabel('Miles Per Gallon (mpg)')

plt.show()
```

# 6.What is the effect of number of cylinders on the mpg ?


```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='cylinders', y='mpg', data=data)

plt.title('Effect of Number of Cylinders on MPG')
plt.xlabel('Number of Cylinders')
plt.ylabel('Miles per Gallon (mpg)')

plt.show()
```

# as we see the no of cylinder increases the mpg decreases

# 8.How much proportion of the total records is occupied by vehicles with 8 cylinders ?

```python
total_records = len(data)

vehicles_with_8_cylinders = len(data[data['cylinders'] == 8])

proportion_8_cylinders = vehicles_with_8_cylinders / total_records

print(f"Proportion of vehicles with 8 cylinders: {proportion_8_cylinders:.2f}")
```
